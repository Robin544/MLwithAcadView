```python
#                                       VISUALIZATION


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Question 1: Download the weather dataset from "https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv".
# a.) Plot atleast 20 visualization for the same dataset with different data points.
# Ensure that the type of visualization is appropriate and follows this chart.

data = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
data.dropna()
data.head()
```

```python
sns.distplot(data["Evaporation"], bins=4)
```

```python
sns.kdeplot(data['Sunshine'], shade = True)
```

```python
sns.distplot(data['MaxTemp'], hist = False)
```

```python
sns.jointplot(x = 'Evaporation', y = 'Rainfall', data = data)
```

```python
sns.boxplot()
```

```python
sns.jointplot(x = data['MinTemp'], y = data['MaxTemp'], kind = 'hex', color = 'k')
```

```python
sns.jointplot(x = data['MinTemp'], y = data['MaxTemp'], data = data, kind = 'kde')
```

```python
sns.stripplot(x = data['MaxTemp'], y = data['MinTemp'], data = data)
```

```python
sns.swarmplot(x = data['MaxTemp'], y = data['Rainfall'], data = data)
```

```python
sns.boxplot(data['MaxTemp'])
```

```python
sns.countplot(data['MinTemp'])
```

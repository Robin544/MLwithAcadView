```python
# QUESTION 3:
# Univariate Analysis - Plot the univariate graphs and find the outliers.
# Explain how you have handled the outliers by transformation (optional).

plt.figure(figsize=(10,6))
sns.distplot(data["price"], hist = True, kde = False)

print("Here, there is no missing value present in the dataset if there is any missing value present in the whole row or whole column then we will remove it as a whole by dropna() function or will find with imputation. In imputation we find out the mean, median of data which is numerical. We can find out imputation on any row we want or the same row and put the value of mean in the missing value place. We use mode as imputation for the categorical data and fill in the place by fillna() function.")
```

```python
# QUESTION 4:
# Bivariate Analysis - Plot the bivariate graphs and find the relationship between the variables.
# Explain your findings and inference of the graphs.

plt.figure(figsize = (15, 12))
sns.heatmap(data.corr(), annot=True)
```

```python
# price is independent on the year it is build. So no inference can be made
plt.figure(figsize = (11, 7))
sns.jointplot(x = data["price"], y = data['yr_built'], kind = "scatter")
```

```python
#Inference : We get a randomised data with some outliers where price is not corelated with the number of bathrooms 
#in the house hence predictions made using this data will be wrong.
plt.figure(figsize = (11, 7))
sns.jointplot(data = data, x = "bathrooms", y = "price", kind = "scatter")
```

```python
#Inference : price is not perfectly corelated with the number of bedrooms in the house, hence predictions cannot be made.
sns.jointplot(data = data, x = "bedrooms", y = "price", kind = "scatter")
```
